```python
df = pd.read_csv(r'C:\Users\deena\OneDrive\Desktop\data.csv')
```

```python
print(df.head())
```

```python
# Melt data for users
df_melted_users = pd.melt(df, id_vars=['city_name'], 
                          value_vars=['Active Users Before 5G', 'Active Users After 5G'],
                          var_name='TimePeriod', value_name='Active Users')

# Add a period column to differentiate before and after 5G
df_melted_users['Period'] = df_melted_users['TimePeriod'].apply(lambda x: 'Before 5G' if 'Before' in x else 'After 5G')

# Repeat the same for ARPU
df_melted_arpu = pd.melt(df, id_vars=['city_name'], 
                          value_vars=['ARPU Before 5G', 'ARPU After 5G'],
                          var_name='TimePeriod', value_name='ARPU')

df_melted_arpu['Period'] = df_melted_arpu['TimePeriod'].apply(lambda x: 'Before 5G' if 'Before' in x else 'After 5G')

# Repeat the same for revenue
df_melted_revenue = pd.melt(df, id_vars=['city_name'], 
                            value_vars=['Revenue Before 5G', 'Revenue After 5G'],
                            var_name='TimePeriod', value_name='Revenue')

df_melted_revenue['Period'] = df_melted_revenue['TimePeriod'].apply(lambda x: 'Before 5G' if 'Before' in x else 'After 5G')

# Repeat the same for unsubscribed users
df_melted_unsubscribed = pd.melt(df, id_vars=['city_name'], 
                                 value_vars=['Unsubscribed Users Before 5G', 'Unsubscribed Users After 5G'],
                                 var_name='TimePeriod', value_name='Unsubscribed Users')

df_melted_unsubscribed['Period'] = df_melted_unsubscribed['TimePeriod'].apply(lambda x: 'Before 5G' if 'Before' in x else 'After 5G')
```

```python
# Perform ANOVA for Active Users
model_active_users = ols('Q("Active Users") ~ C(Period)', data=df_melted_users).fit()
anova_table_active_users = sm.stats.anova_lm(model_active_users, typ=2)
print(anova_table_active_users)
```

```python
# Perform ANOVA for ARPU
model_arpu = ols('Q("ARPU") ~ C(Period)', data=df_melted_arpu).fit()
anova_table_arpu = sm.stats.anova_lm(model_arpu, typ=2)
print(anova_table_arpu)
```

```python
df_melted_arpu.boxplot(column='ARPU', by='Period', grid=False)
plt.title('ARPU Before and After 5G Introduction')
plt.suptitle('')
plt.xlabel('Time Period')
plt.ylabel('ARPU')
plt.show()
```

```python
# Perform ANOVA for Revenue
model_revenue = ols('Q("Revenue") ~ C(Period)', data=df_melted_revenue).fit()
anova_table_revenue = sm.stats.anova_lm(model_revenue, typ=2)
print(anova_table_revenue)
```

```python
# Perform ANOVA for Unsubscribed Users
model_unsubscribed = ols('Q("Unsubscribed Users") ~ C(Period)', data=df_melted_unsubscribed).fit()
anova_table_unsubscribed = sm.stats.anova_lm(model_unsubscribed, typ=2)
print(anova_table_unsubscribed)
```
